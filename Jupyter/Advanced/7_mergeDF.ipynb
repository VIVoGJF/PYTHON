```python
import numpy as np
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
df_us = df[df.job_country == 'United States'].copy()

df_us['job_posted_month'] = pd.to_datetime(df.job_posted_date).dt.strftime('%B')

df_us_pivot = df_us.pivot_table(
    columns='job_title_short',
    index='job_posted_month',
    aggfunc='size'
)

df_us_pivot.reset_index(inplace=True)
df_us_pivot['m_no'] = pd.to_datetime(df_us_pivot.job_posted_month, format='%B').dt.month
df_us_pivot.sort_values('m_no',inplace=True)
df_us_pivot.set_index('job_posted_month', inplace=True)
df_us_pivot.drop(columns='m_no',inplace=True)
df_us_pivot
```

```python
df_us_soft_pivot = pd.read_csv("http://lukeb.co/software_csv", index_col='job_posted_month')
df_us_soft_pivot
```

```python
df_us_merge = df_us_pivot.merge(df_us_soft_pivot, on='job_posted_month')

df_us_merge
```

```python
top_5 = df_us_merge.sum().sort_values(ascending=False).head(5).index.tolist()

df_us_merge[top_5].plot(kind='line')
plt.title('Monthly job Postings for Top Data Jobs in US')
plt.ylabel('Job Count')
plt.xlabel('2023')
plt.ylim(0,20000)
plt.legend()
plt.show()
```
