```python
import matplotlib.pyplot as plt
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
df['job_posted_date']=pd.to_datetime(df.job_posted_date)
```

```python
df.info()
```

```python
df['job_posted_month']=df.job_posted_date.dt.month
```

```python
moncnt = df.job_posted_month.value_counts().sort_index()

moncnt
```

```python
plt.plot(moncnt.index, moncnt)
```

```python
job1 = df['job_title_short'].value_counts()
```

```python
plt.bar(job1.index, job1)
plt.title('Posting by Job Title')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
job2 = df['job_title_short'].value_counts().sort_values(ascending=True)
```
